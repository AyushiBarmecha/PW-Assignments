**Q6. Explain the concept of generators in Python and how they are defined.**

**Answer:**

A generator function in Python is defined like a normal function, but whenever it needs to generate a value, it does so with the yield keyword rather than return. If the body of a def contains yield, the function automatically becomes a Python generator function.

In Python, we can create a generator function by simply using the def keyword and the yield keyword. Generator objects are used either by calling the next method of the generator object or using the generator object in a “for in” loop.  Generator is derived from Iterator is derived from Iterable is derived from the base object. Calling iter() on an iterator or a generator returns itself.

```python
def simpleGeneratorFun():        # A generator function
    yield 1
    yield 2
    yield 3

for value in simpleGeneratorFun():
    print(value)



generator_exp = (i * 5 for i in range(5) if i%2==0)     # generator expression

for i in generator_exp:
    print(i)


def simpleGeneratorFun():      # generator object
    yield 1
    yield 2
    yield 3

x = simpleGeneratorFun()    # x is a generator object
```

**Q7. What are the advantages of using generators over regular functions?**

**Answer:**

Generators are a special type of iterator created using the def keyword followed by the yield statement. Unlike regular functions that use return to provide a single value, generator functions use yield to produce a sequence of values on demand.

This allows generators to be memory-efficient and capable of representing infinite streams of data. When a generator function is called, it returns a generator object that can be iterated over to retrieve the values.

**Advantages of generators include:**

1. **Lazy evaluation:** Values are computed only when explicitly requested, leading to efficient memory usage.
2. **Memory efficiency:** Generators produce values on the fly, avoiding the need to store all elements in memory.
3. **Infinite sequences:** Generators can represent infinite streams of data without consuming excessive memory.
4. **Pipelining:** Generators enable a series of operations to be chained together efficiently.
5. **Readability and maintainability:** Generator functions encapsulate complex iteration logic, enhancing code clarity.
6. **Usage with Functions:**
Generator expressions can be used directly as arguments to functions that accept iterables. They are especially useful with reduction functions like sum(), min(), max().

**Difference Between Generator and Regular Function**
1. In generator, there are one or more yield functions, whereas, in Regular functions, there is only one function
2. When the generator is called, the regular function pauses its execution, and the call is transferred to the generator.
3. Local variables and their states are remembered between successive calls.
4. When the generator is terminated, StopIteration is called automatically on further calls.
5. Generator use 'Yeild'. It is responsible for controlling the flow of the generator function. After returning the value from yield, it pauses the execution by saving the states. whereas Regular functions  use 'Return'. It returns the value and terminates the function.

**Q10. What is the difference between map(), reduce(), and filter() functions in Python?**

**Answer:**

* **Map()** is used to apply a function on every item in an array and returns the new array.

* **Filter()** is used to create a new array from an existing one, containing only those items that satisfy a condition specified in a function.

* **Reduce()** is used to apply a function against an accumulator and each element in the array, with the aim of reducing the array to a single output value.

**Difference between map(), reduce(), and filter()**
1. `'Map'` operation in programming refers to the transformation of data from one form into another. The `'Reduce'` operation is a method for combining the elements of a list into a single result, while `'Filter'` is a method for selecting a subset of items from a list based on a particular condition.
2. `'Map'` applies a given function to each item of an iterable (like a list or set) and returns a list of the results. While, `'Filter'` constructs a list from elements of an iterable for which a function returns true. Whereas, `'Reduce'` applies a rolling computation to sequential pairs of values in a list and returns a single result.
3. `'Map'` is used to apply a function on every item in an array and returns the new array. Whereas, `'Filter'` is used to create a new array from an existing one, containing only those items that satisfy a condition specified in a function. Whereas, `'Reduce'` is used to apply a function against an accumulator and each element in the array, with the aim of reducing the array to a single output value.
4. `'Filter'` extracts elements from an iterable (like lists) for which a function returns true. Whereas, `'Map'`applies a function to all items in an input list. Whereas, `'Reduce'` applies a function of two arguments cumulatively to the items of an iterable to reduce the iterable to a single output.

5. `'Map'` is transformation of one form of data into another. `'Reduce'` is combination of list elements into a single result. `'Filter'` is selection of list subset based on a certain condition.


```python
# using map()

numbers = [1, 2, 3, 4, 5]
double_numbers = list(map(lambda x: x * 2, numbers))
print(double_numbers)


# using filter()

numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)


# using reduce()

from functools import reduce
numbers = [1, 2, 3, 4, 5]
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```

**Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def prime_number(num):
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True

number = 13
print(f"Is {number} a prime number? {prime_number (number)}")
```

**Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
def fibonacci():
    x, y = 0, 1
    while True:
        yield x
        x, y = y, x + y

# Accept input from the user
n = int(input("Input the number of Fibonacci numbers you want to generate? "))

print("Number of first ",n,"Fibonacci numbers:")
fib = fibonacci()
for _ in range(n):
    print(next(fib),end=" ")
```

**Q7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Yield each line without the newline character
    except FileNotFoundError:
        print(f"The file at {file_path} does not exist.")
    except IOError as e:
        print(f"An error occurred while reading the file: {e}")

# Example usage:
file_path = 'example.txt'  # Replace with your file path
for line in read_file_line_by_line(file_path):
    print(line)
```

**Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.**

**Write a Python program using lambda and map.**

```python
# Define the input data
orders = [
    ("34587", 4, 40.95),
    ("98762", 5, 56.80),
    ("77226", 3, 32.95),
    ("88112", 3, 24.99)
]

# Define the lambda() to calculate adjusted total cost
adjusted_cost = lambda order: (
    order[0],  # Order number
    max(order[1] * order[2], 100) + (10 if order[1] * order[2] < 100 else 0)  # Adjusted cost
)

# Use map() to apply the lambda() to each order
result = list(map(adjusted_cost, orders))

# Print the result
print(result)
```
